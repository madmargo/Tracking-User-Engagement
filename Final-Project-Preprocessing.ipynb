```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
q2_2021_non_paid = pd.read_csv('minutes-watched-2021-paid-0.csv')
q2_2021_paid = pd.read_csv('minutes-watched-2021-paid-1.csv')
q2_2022_non_paid = pd.read_csv('minutes-watched-2022-paid-0.csv')
q2_2022_paid = pd.read_csv('minutes-watched-2022-paid-1.csv')
```

```python
q2_2021_non_paid.head()
```

```python
q2_2021_paid.head()
```

```python
q2_2022_non_paid.head()
```

```python
q2_2022_paid.head()
```

```python
fig, axs = plt.subplots(2, 2, figsize = (9,9))
axs[0,0].plot(q2_2021_non_paid['minutes_watched'], 'r')
axs[0,0].set_title('Q2 2021 Non-Paid')
axs[0,1].plot(q2_2021_paid['minutes_watched'], 'g')
axs[0,1].set_title('Q2 2021 Paid')
axs[1,0].plot(q2_2022_non_paid['minutes_watched'], 'b')
axs[1,0].set_title('Q2 2022 Non-Paid')
axs[1,1].plot(q2_2022_paid['minutes_watched'], 'k')
axs[1,1].set_title('Q2 2022 Paid')
plt.show()
```

```python
print(q2_2021_non_paid['minutes_watched'].quantile(0.99))
q2_2021_non_paid_no_outliers = q2_2021_non_paid[q2_2021_non_paid['minutes_watched'] < 167.0847]
print(q2_2021_non_paid_no_outliers)
q2_2021_non_paid_no_outliers.to_csv('minutes-watched-2021-paid-0-no-outliers.csv')
```

```python
print(q2_2021_paid['minutes_watched'].quantile(0.99))
q2_2021_paid_no_outliers = q2_2021_paid[q2_2021_paid['minutes_watched'] < 3390.4828]
print(q2_2021_paid_no_outliers)
q2_2021_paid_no_outliers.to_csv('minutes-watched-2021-paid-1-no-outliers.csv')
```

```python
print(q2_2022_non_paid['minutes_watched'].quantile(0.99))
q2_2022_non_paid_no_outliers = q2_2022_non_paid[q2_2022_non_paid['minutes_watched'] < 170.2630]
print(q2_2022_non_paid_no_outliers)
q2_2022_non_paid_no_outliers.to_csv('minutes-watched-2022-paid-0-no-outliers.csv')
```

```python
print(q2_2022_paid['minutes_watched'].quantile(0.99))
q2_2022_paid_no_outliers = q2_2022_paid[q2_2022_paid['minutes_watched'] < 2505.3450]
print(q2_2022_paid_no_outliers)
q2_2022_paid_no_outliers.to_csv('minutes-watched-2022-paid-1-no-outliers.csv')
```

```python
fig, axs = plt.subplots(2, 2, figsize = (9,9))
axs[0,0].plot(q2_2021_non_paid_no_outliers['minutes_watched'], 'r')
axs[0,0].set_title('Q2 2021 Non-Paid')
axs[0,1].plot(q2_2021_paid_no_outliers['minutes_watched'], 'g')
axs[0,1].set_title('Q2 2021 Paid')
axs[1,0].plot(q2_2022_non_paid_no_outliers['minutes_watched'], 'b')
axs[1,0].set_title('Q2 2022 Non-Paid')
axs[1,1].plot(q2_2022_paid_no_outliers['minutes_watched'], 'k')
axs[1,1].set_title('Q2 2022 Paid')
plt.show()
```
